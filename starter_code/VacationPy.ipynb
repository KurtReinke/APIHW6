```python
#Store into variable named hotel_df.
#Add a "Hotel Name" column to the DataFrame.
#Set parameters to search for hotels with 5000 meters.
#Hit the Google Places API for each city's coordinates.
#Store the first Hotel result into the DataFrame.
#Plot markers on top of the heatmap.

#Create a list then a df


#https://maps.googleapis.com/maps/api/place/nearbysearch/json
  #?location=-33.8670522,151.1957362
 # &radius=500
 # &types=food
  #&name=harbour
 # &key=YOUR_API_KEY 
#attach url to pull data from google site
#url2="https://maps.googleapis.com/maps/api/place/nearbysearch/json"
#query_url2 = f"{url2}appid={g_key}&units={hotel}&q="
#response1=requests.get(url2)
#response1_json=response1.json

#limits={"radius":5000,
      #"types":"hotel",
      #"key":g_key}

#url2="https://maps.googleapis.com/maps/api/place/nearbysearch/json"
#response1=requests.get(url2, limits)
#response1_json=response1.json

hotel_df=nonull_df
list1=[]

#Create loop for google data
#define key and url
#loc lat and long
#filter out raidus <5000
for x in range(len(hotel_df["City"])):
    lat=hotel_df.loc[x]["Lat"]
    lng=hotel_df.loc[x]["Lng"]
    
#define lat and long of vacation spots    
    vacationcities=f"({Lat},{Lng})"
    
    limits={"radius":5000,
            "location":vacationcities,
            #"types":"hotel",
            "key":g_key}
    
#call url and response json
    url2="https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response1=requests.get(url2, limits)
    response1_json=response1.json()
    
#complete loop with try and except    
    try:
        list1.append(response1_json["results"][0]["name"])
        #list1.append("hotel")
    except:
        list1.append("Hotel Not Found")

        
#append list to df
hotel_df["Hotel Name"]=list1
hotel_df
#filter results
```
